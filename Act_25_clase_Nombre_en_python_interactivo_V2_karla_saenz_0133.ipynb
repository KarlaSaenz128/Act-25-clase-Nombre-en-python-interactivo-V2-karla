```python
class Sensor:
    def __init__(self, tipo, valor, unidad):
        self.tipo = tipo
        self.valor = valor
        self.unidad = unidad

    def leerdatos(self):
        self.valor = float(input(f"Nuevo valor para {self.tipo}: "))

    def mostrardatos(self):
        print(f"Sensor {self.tipo}: {self.valor} {self.unidad}")

# --- Creación de Objetos ---
sensor1 = Sensor("Temperatura", 25.0, "°C")
sensor2 = Sensor("Humedad", 60, "%")

# --- Prueba de Funciones ---
sensor1.mostrardatos()
sensor1.leerdatos()
sensor1.mostrardatos()

sensor2.mostrardatos()
print(" elaborado por Karla Saenz 0133")
```
